### **TIMES DATA**

```python
import pandas as pd
data5 = pd.read_csv("/content/sample_data/THP/timesData.csv")
data5
```

```python
year5 = data5['year']
year5
```

les années sont entre 2011 et 2016

```python
strasbourg_times_data = data5[data5['university_name'] == 'University of Strasbourg']

# Display the data
strasbourg_times_data[['world_rank', 'year']]
```

les Rankings de Strasbourg



```python
# Filtrage data pour 2016
times_data_2016 = data5[data5['year'] == 2016]

# Convertir le world_rank en numerique quand c'est possible
times_data_2016['world_rank'] = pd.to_numeric(times_data_2016['world_rank'], errors='coerce')

# avoir le top 20 universities in 2016
top_20_universities_2016 = times_data_2016.sort_values(by='world_rank').head(20)

# affichage top 20 universities in 2016
top_20_universities_2016[['world_rank', 'university_name', 'country']]
```

```python
# Count du nombres d'universités US dans le top 20
num_us_universities = top_20_universities_2016[top_20_universities_2016['country'] == 'United States of America'].shape[0]

# Calcul des proportions American universities in the top 20
proportion_us_universities = num_us_universities / 20

proportion_us_universities * 100
```

il y'a 70% d'universités américaines dans le top 20

# **CALCUL DU RATIO CLASSEMENT SUPERIEUR ET TIMES**

```python
# Filtre times_data for the year 2011
times_data_2011 = data5[data5['year'] == 2011]

# Count the number of universities dans chaque pays
universities_per_country_times = times_data_2011['country'].value_counts()

# Conversion de la Serie en DataFrame
universities_per_country_times_df = universities_per_country_times.reset_index()
universities_per_country_times_df.columns = ['country', 'num_universities']

# Display
universities_per_country_times_df.head()
```

```python
# on additionne les data de chaque country
expenditure= data1.groupby('country')['2011'].sum().reset_index()

# Merge des dataframes
merged_data_agg = pd.merge(universities_per_country_times_df, expenditure, left_on='country', right_on='country', how='inner')

# Calcul du ratio
merged_data_agg['ratio'] = merged_data_agg['2011'] / merged_data_agg['num_universities']

# filtrage de la data by ratio
sorted_data_agg = merged_data_agg.sort_values(by='ratio')

# Display
sorted_data_agg[['country', 'ratio']]
```
